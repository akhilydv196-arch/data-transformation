**3.What is the difference between Normalization and Standardization?**

Both Normalization and Standardization are feature scaling techniques used to bring different variables into a comparable range. While they sound similar, they treat the data distribution quite differently.

1.Normalization (Min-Max Scaling)

Normalization rescales the data into a fixed range, usually [0, 1]. It is most useful when you know the distribution of your data does not follow a Gaussian (Normal) distribution.

Drawback: It is highly sensitive to outliers.

2.Standarization(z-score scaling )

Standardization centers the data around a mean of 0 with a standard deviation of 1. Unlike normalization, it does not have a bounding range (like 0 to 1).

Advantage: It is much more robust to outliers
